```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, 
                                y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg_clf = LogisticRegression()
log_reg_clf.fit(X_train,y_train)
log_reg_clf.score(X_test , y_test)
```

```python
from sklearn.svm import SVC
svc_clf = SVC()
svc_clf.fit(X_train,y_train)
svc_clf.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_for_clf = RandomForestClassifier()
ran_for_clf.fit(X_train,y_train)
ran_for_clf.score(X_test,y_test)
ran_for_clf.score(X_train,y_train)
```

```python
ran_for_clf.score(X_train,y_train)
```

```python
def getscore(model,X_train,y_train):
    model.fit(X_train,y_train)
    return model.score(X_train,y_train)
```

```python
getscore(RandomForestClassifier(),X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(RandomForestClassifier(),X_train,y_train)
```
